# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular data visualization library in Python that provides a wide range of functionalities for creating static, interactive, and animated visualizations. It is widely used in various fields, such as data analysis, scientific research, machine learning, and more. Matplotlib is capable of generating high-quality charts, plots, histograms, 3D visualizations, and much more.

Here are five types of plots that can be created using the Pyplot module of Matplotlib:

Line Plot: Line plots are used to visualize data points connected with straight lines, often used to show the trend or relationship between two variables.

Scatter Plot: Scatter plots display individual data points as dots on a 2D plane, with one variable on the x-axis and another on the y-axis, showing the distribution and correlation between the two variables.

Bar Plot: Bar plots represent categorical data with rectangular bars, where the length of each bar is proportional to the values they represent. They are often used to compare data across different categories.

Histogram: Histograms are used to display the distribution of a continuous numerical variable, dividing the data into bins and representing the frequency of data points falling into each bin.

Pie Chart: Pie charts display data as a circular graph, where each category is represented as a slice, and the size of each slice is proportional to the percentage of data it represents in the whole.

The subplot() function in Matplotlib is used to create multiple plots within the same figure. It helps in arranging multiple plots in a grid layout, making it easier to visualize and compare different datasets or plot types side by side. The subplot() function takes three integer arguments: the number of rows, the number of columns, and the plot index. The plot index ranges from 1 to the total number of subplots (rows * columns).

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid layout for four subplots
plt.subplot(2, 2, 1)  # First subplot
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)  # Second subplot
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(3, 3, 3)  # Third subplot
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(3, 3, 4)  # Fourth subplot
plt.plot(x4, y4)
plt.title('Line 4')

# Adjust the layout to prevent overlapping of titles
plt.tight_layout()

# Show the plot
plt.show()
```

A bar plot, also known as a bar chart, is a data visualization that represents categorical data with rectangular bars. The length of each bar is proportional to the value it represents. Bar plots are commonly used to compare and display the relative magnitude or frequency of different categories, making them suitable for showing discrete data distributions.

The primary uses of bar plots are as follows:

Comparison: Bar plots are effective for comparing the values of different categories side by side, allowing quick visual identification of the largest or smallest values.

Distribution: They can be used to visualize the distribution of data across different categories, showing the spread and variation.

Trends: Bar plots can be used to show trends or changes in data over time or between different groups.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company , profit)
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a data visualization technique that provides a graphical summary of the distribution of a dataset. It displays the five-number summary of the data, which includes the minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. The "box" in the plot represents the interquartile range (IQR), which is the range between Q1 and Q3. The "whiskers" extend from the box to the minimum and maximum values within a certain distance from the quartiles, and any data points outside this range are considered outliers and represented as individual points.

Box plots are used for the following purposes:

Visualizing Distribution: Box plots provide a concise visual summary of the distribution of a dataset, including information about central tendency (median) and spread (IQR).

Identifying Outliers: They help in identifying potential outliers in the data, which are points lying significantly outside the main range of the data.

Comparing Distributions: Box plots are useful for comparing the distributions of multiple datasets side by side, making it easier to spot differences in their central tendency and variability.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
plt.boxplot(data , labels =["box1", "box2"])
```
